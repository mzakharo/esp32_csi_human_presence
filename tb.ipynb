```python
%load_ext autoreload
%autoreload 2
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
#mat = scipy.io.loadmat('bar.mat')
mat = scipy.io.loadmat('capture3.mat')
#mat = scipy.io.loadmat('capture4.mat')

CSI = np.abs(mat['CSI'])
print(CSI)
```

```python
plt.plot(CSI[:, 15])
```

```python
conf = []
import mdet

detector = mdet.CSIMagnitudeDetector()

for i in range(CSI.shape[0]):
     confidence = list(detector.detect_presence(CSI[i, :]))
     confidence.append(detector.var)
     confidence.append(detector.cor)
     conf.append(confidence)
plt.plot(conf, label= ['mot', 'conf', 'var', 'corr'])

plt.legend(loc='upper left')
```
